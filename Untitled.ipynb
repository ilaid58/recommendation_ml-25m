## 1- get features of movies
   The column 'genre' of the dataframe `movies` has differents features of movies that we want to extract
   'genre' columns can have a value like : `Adventure|Animation|Children|Comedy|Fantasy`. Each word separated by | are a feature of movies

```python
# 
get_feat = get_features(movies)
genre_feat = get_feat.on_moviescsv()
```

## 2- define feature
Above, we extract features on 'genre' column for `movies dataframe`.
- `genre()` function create dataframe with feature genre columns as values binary number related with each movieId.

- `tag()` function create dataframe with tag name as features and relevance value of df related to each movieId.

```python
def_feat = define_feature(genre_feat, movies, g_scores)
df_genre = def_feat.genre()

df_tag = def_feat.tag()
```
